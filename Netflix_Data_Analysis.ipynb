```python
# Categorizing Vote_Average Column

We would cut the Vote_Average values and make 4 categories:

- popular
- average
- below_avg
- not_popular

to describe it more using the categorize_col() function provided above.
```

```python
def categorize_col(df, col, labels):
    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]

    df[col] = pd.cut(df[col], bins=edges, labels=labels, duplicates='drop')

    return df
```

```python
labels=['not popular','below avg','avg','popular']
categorize_col(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```
